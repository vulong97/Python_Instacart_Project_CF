## If- statement with user-defined functions

```python
#Defining the function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts()
```

```python
#Checking maximum price in the subset 'df'
df['prices'].max()
```

### Creating new column 'Busiest days'

```python
df_ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

#### Busiest days are 0 and 1 (Saturday and Sunday), least busy days are 3 and 4 (Tuesday and Wednesday)

```python
#Creating a loop for the 'order_day_of_week' values
days = []

for value in df_ords_prods_merged['order_day_of_week']:
  if value == 0 or value == 1:
    days.append('Busiest days')
  elif value == 3 or value == 4:
    days.append('Slowest days')
  else:
    days.append('Regularly busy')
```

```python
#Creating the column 'busiest_days' and adding to the dataframe
df_ords_prods_merged['busiest_days'] = days
```

```python
#Checking value count
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

### Creating new column 'busiest_period_of_day'

```python
#Checking the frequency of the values in 'order_hour_of_day'
count_of_orders = df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
count_of_orders
```

```python
#Searching for percentile of the count of values
list = list(df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).quantile([0.75, 0.5, 0.25]))
```

```python
list
```

##### Based on the percentiles, we can divide the hours into 3 groups:
##### Group 1 (Fewest orders - Under 272562 orders): 3, 4, 2, 5, 1, 0
##### Group 2 (Average orders - Between 272562 and 2556034.5): 6, 23, 22, 21, 7, 20,19,18,8,17,9,16
##### Group 3 (Most orders - Over 2556034.5): 12,13,15,14,11,10

```python
#Creating a loop for 'order_hour_of_day' values
hours = []

for value in df_ords_prods_merged['order_hour_of_day']:
  if value in [10, 11, 14, 15, 13, 12]:
    hours.append('Most orders')
  elif value in [0, 1, 5, 2, 4, 3]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
#Adding the column 'busiest_period_of_day' to the dataframe
df_ords_prods_merged['busiest_period_of_day'] = hours
```
