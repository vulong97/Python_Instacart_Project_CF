### Step 4

#### Create a loyalty flag for existing customers using the transform() and loc() functions.

```python
#Split the data into groups based on the 'user_id' column. Apply the transform() function on the 'order_number' column to generate the max orders for each user.
df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

### Step 5

#### Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer).

```python
#Performing multiple aggregations ('min', 'max', 'mean', 'sum')
df_ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max', 'sum']})
```

### Step 6

#### Create a spending flag for each user based on the average price across all their orders using the following criteria:
#### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
#### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#Split the data into groups based on the 'user_id' column. Apply the transform() function on the 'prices' column to generate the average prices for each user.
df_ords_prods_merged['avg_price'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

### Step 7

#### Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
#### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#Split the data into groups based on the 'user_id' column. Apply the transform() function on the 'days_since_prior_order' column to generate the median order frequency for each user.
df_ords_prods_merged['median_freq'] = df_ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_freq'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['median_freq'] > 10) & (df_ords_prods_merged['median_freq'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_freq'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```
